```python
#importing necessary libraries
import pandas as pd
import numpy as np
from sklearn import linear_model
% matplotlib inline
import seaborn as sb
```

```python
data = pd.read_csv('UCD1999-2015CSVDATA.csv')
data['Calculation'] = data['Deaths'] / data['Population'] * 100000
sorteddf = data.sort(['Calculation'], ascending=False)
sorteddf
```

```python
inputrace = input('Enter the race you would like to visualize (American Indian or Alaska Native, Asian or Pacific Islander, Black or African American, or White): '+ "\n" )
```

```python
inputgender = input('Enter the gender you would like to visualize (Male or Female): '+ "\n" )
```

```python
inputdisease = input('Enter the the disease you want to visualize:'+ "\n" + "Diseases of the circulatory system" + "\n" + "Neoplasms" + "\n" + "Diseases of the respiratory system " + "\n" + "Diseases of the nervous system  " + "\n" + "Mental and behavioural disorders" + "\n" + "Certain conditions originating in the perinatal period" + "\n" + "Diseases of the genitourinary system" + "\n" + "Diseases of the genitourinary system" + "\n" + "Endocrine, nutritional and metabolic diseases" + "\n" + "External causes of morbidity and mortality " + "\n" + "Certain infectious and parasitic diseases" + "\n" + "Diseases of the digestive system" + "\n" + "Symptoms, signs and abnormal clinical and laboratory findings, not elsewhere classified" + "\n" + "Congenital malformations, deformations and chromosomal abnormalities" + "\n" + "Diseases of the musculoskeletal system and connective tissue" + "\n" + "Diseases of the skin and subcutaneous tissue" + "\n" + "Diseases of the blood and blood-forming organs and certain disorders involving the immune mechanism" + "\n" + "Pregnancy, childbirth and the puerperium" + "\n" + "Diseases of the ear and mastoid process" + "\n" + "Diseases of the eye and adnexa" + "\n" + "Codes for special purposes" + "\n"  )
```

```python
uniquecategory = sorteddf[(sorteddf['Race'] == inputrace) & (sorteddf['Gender'] == inputgender) & (sorteddf['ICDChapter'] == inputdisease)]
uniquecategory.sort(['Age'], ascending=False)
uniquecategory
```

## Visualizing your prefered disease in Seaborn

```python
sb.lmplot(x ='Age', y = 'Calculation', data = uniquecategory, hue = 'ICDChapter', ci = 0)
```

## Visualizing your selected disease in Matplotlib

```python
plt = uniquecategory.reset_index().plot.line(x = 'Age', label = inputdisease, y = 'Calculation')

#plt.gca().invert_xaxis()
```
